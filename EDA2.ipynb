# 2. in which year star war is not famoused in licensed 

```python
licensed
```

```python
licensed[["set_num","year","parent_theme"]]
```

```python
licensed["count"]=1
```

```python
licensed=licensed.groupby(["year","parent_theme"]).sum().reset_index()
licensed                   
```

```python
year_wise=licensed.sort_values("count",ascending=False).drop_duplicates(["year"])
year_wise
```

```python
year_wise=year_wise.sort_values("year")
year_wise
```
